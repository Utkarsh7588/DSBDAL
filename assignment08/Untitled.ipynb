```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = sns.load_dataset('titanic')
titanic_data
```

```python
sns.histplot(data=titanic_data, x='fare', kde=True)
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data=titanic_data, x='sex', hue='survived')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(data=titanic_data, x='class', y='fare')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
numeric_columns = titanic_data.select_dtypes(include=['int64', 'float64']).columns

# Compute correlation matrix
correlation_matrix = titanic_data[numeric_columns].corr()

# Plot the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python

sns.pairplot(data=titanic_data, vars=['age', 'fare', 'pclass'])
plt.title('Pairplot of Numeric Variables')
plt.show()
```
